```python
import os
import json
import csv
import numpy as np

json_path = r'E:\AInotes\자세교정\전처리\RNN데이터 전처리\keypoint\label_data_mlkit'
out_path = r'E:\AInotes\자세교정\전처리\RNN데이터 전처리\sequence_label'

for roots, dirs, files in os.walk(json_path):
    for file in files:
        data_arr = []
        label_arr = []
        keypoint_arr = []
        if file.endswith('.json'):
            file_path = os.path.join(roots, file)
            root, dir = os.path.split(roots)
            with open(file_path, 'r', encoding='utf-8') as json_file:
                data_list = json.load(json_file)
                flag = 0                
                for data in data_list:
                    value = data['filename']
                    label_value = value[6:9]
                    if label_value not in label_arr:
                        label_arr.append(label_value)                   
                    
                    if label_value == flag or flag==0:
                        pts_arr = []
                        for key, value in data['pts'].items():
                            pts_arr.append(value)
                        keypoint_arr.append(pts_arr)
                    else:
                        data_arr.append(keypoint_arr)
                        keypoint_arr = []
                        pts_arr = []
                        for key, value in data['pts'].items():
                            pts_arr.append(value)
                        keypoint_arr.append(pts_arr)
                    flag = label_value
            if keypoint_arr:
                data_arr.append(keypoint_arr)
            # 데이터를 딕셔너리로 구성
            data_to_save = {"sequences": data_arr, "labels": label_arr}
            
            # JSON 파일로 저장
            with open(out_path + f"\\{dir}_{file[6]}sequences_label.json", "w") as json_file:
                json.dump(data_to_save, json_file, indent=2)
            
            print('done')

```

```python
dataset_path = r'E:\AInotes\자세교정\전처리\RNN데이터 전처리\sequence_label'
# JSON 파일에서 데이터 불러오기

sequence_data = []
label_data = []

for roots, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith('.json'):
            file_path = os.path.join(roots, file)
            print(file_path)
            with open(file_path, "r") as json_file:
                loaded_data = json.load(json_file)

                # 불러온 데이터 확인
                sequences_loaded = loaded_data["sequences"]
                labels_loaded = loaded_data["labels"]

                # RNN에 사용할 데이터 형식으로 extend 저장
                sequence_data.extend(sequences_loaded)
                label_data.extend(labels_loaded)
                # 시퀀스의 개수 확인
                num_sequences = len(sequences_loaded)
                    
                print("시퀀스의 개수:", num_sequences)
                print("라벨의 개수:", len(labels_loaded))

num_sequence_data = len(sequence_data)

print("시퀀스의 개수:", num_sequence_data)
print("라벨의 개수:", len(label_data))
print(label_data)
```
