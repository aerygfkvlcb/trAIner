```python
import os
import shutil

#folder_list = []

def copy_folders_with_prefix(root_folder, destination_folderC, destination_folderD, last_path, prefix, suffixC, suffixD):
    for parent_folder in os.listdir(root_folder):
        parent_folder_path = os.path.join(root_folder, parent_folder)
        if os.path.isdir(parent_folder_path):
            for sub1_folder in os.listdir(parent_folder_path):
                sub1_folder_path = os.path.join(parent_folder_path, sub1_folder)
                if os.path.isdir(sub1_folder_path):
                    for sub2_folder in os.listdir(sub1_folder_path):
                        sub2_folder_path = os.path.join(sub1_folder_path, sub2_folder)
                        if os.path.isdir(sub2_folder_path):
                            for folder_name in os.listdir(sub2_folder_path):
                                folder_path = os.path.join(sub2_folder_path, folder_name)
                                if os.path.isdir(folder_path) and folder_name.startswith(prefix) and folder_name.endswith(suffixC):
                                    destination_pathC = os.path.join(destination_folderC, last_path)
                                    destination_C = os.path.join(destination_pathC, folder_name)
                                    shutil.copytree(folder_path, destination_C)
                                    print(f"폴더 '{folder_name}'를 '{destination_C}'로 복사했습니다.C")
                                if os.path.isdir(folder_path) and folder_name.startswith(prefix) and folder_name.endswith(suffixD):
                                    destination_pathD = os.path.join(destination_folderD, last_path)
                                    destination_D = os.path.join(destination_pathD, folder_name)
                                    shutil.copytree(folder_path, destination_D)
                                    print(f"폴더 '{folder_name}'를 '{destination_D}'로 복사했습니다.D")

#def copy_folders(folder_list, destination_folder):
#    for folder_path in folder_list:
#          # 각 폴더를 대상 폴더에 복사
#        shutil.copytree(folder_path, destination_folder)
#        print(f"폴더 '{folder_path}'를 '{destination_folder}'로 복사했습니다.")

path_root = 'D:/AI/dataset/fitnessPose/Training/dataset'
path_face = 'D:/AI/dataset_sort/training/face'
path_side = 'D:/AI/dataset_sort/training/side'
path_exercise = 'StepBackwardDynamicLunge'
last_path = ['10010']
prefix = ['144']
path_C = os.path.join(path_face, path_exercise)
path_D = os.path.join(path_side, path_exercise)

# 'day1' 폴더 안에서 '001'로 시작하는 폴더를 찾아 원하는 경로로 복사하기
for last_path, prefix in zip(last_path, prefix):    
    copy_folders_with_prefix(path_root, path_C, path_D, last_path, prefix, 'C', 'D')
```
